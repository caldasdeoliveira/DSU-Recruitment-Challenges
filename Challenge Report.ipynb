The [Dow Jones Industrial Average](https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average) (DJIA) is a stock market index that measures the stock performance of 30 large companies listed on stock exchanges in the United States. The DJIA is the sum of the prices of all 30 stocks over the Dow Divisor.

The data provided refers to the first two quarters of 2011, January-March and April-June. During this period the companies that constituted the 30 stocks are available in "dow_jones_index.names", see [here](https://en.wikipedia.org/wiki/Historical_components_of_the_Dow_Jones_Industrial_Average) for a complete list over the years.

Since the goal is to find the best company in which to invest weekly, we will need to look at each stock separately rather than the index, which may still be used as a feature. We will only consider the buying and selling of stocks and won't consider other investment methods or the possibility of shorting. We will nonetheless consider the possibility of [fractional investment](https://www.investopedia.com/terms/f/fractionalshare.asp) by assuming we are buying and selling through a major brokerage firm. This assumption means that stock prices can exceed 100x[1.39](https://www.statista.com/statistics/412794/euro-to-u-s-dollar-annual-average-exchange-rate/)$ and the number of bought stocks doesn't need to be taken into account , otherwise a 70€ stock with an increase of 30\% would be worse than 2*50€ stocks with a 25\% increase.

```python
#add here your imports
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import datetime
import seaborn as sns
```

```python
# load data
PATH_TO_DATA = os.path.join("data","dow_jones_index.data")
data_original = pd.read_csv(PATH_TO_DATA)
data = data_original
```

```python
# converting money variables to floats
money_variables = ['open','high','low','close','next_weeks_open','next_weeks_close']

for v in money_variables:
    data[v]=data[v].str[1:].astype(float)
    
data['date'] = pd.to_datetime(data['date'])
data = data.set_index('date')
data['stock']= data['stock'].astype('category')
```

```python
data.isna().sum()
```

```python
# fill NAN values with zero
data['percent_change_volume_over_last_wk'] = data['percent_change_volume_over_last_wk'].fillna(0)
data['previous_weeks_volume'] = data['previous_weeks_volume'].fillna(method='bfill')
```

```python
# Dealing with target leakage

# We do not include "next_weeks_open" as when we decide to buy we can see the current price of the stocks.
# Since we are not told when the decision must be made I am assuming it is made as soon as the next week starts
future_columns = ["next_weeks_close","percent_change_next_weeks_price", "percent_return_next_dividend"]

Y = data[["stock","percent_change_next_weeks_price"]]
temp_Y = data[["stock","next_weeks_close"]]
data= data.drop(columns=future_columns)
```

```python
# divide data by company

data_by_stock = {}
stock_list = []
for stock, d in data.groupby('stock'):
    stock_list.append(stock)
    d=d.drop(columns='stock')
    data_by_stock[stock] = d
Y_dict = {}
for stock, d in Y.groupby('stock'):
    Y_dict[stock] = d["percent_change_next_weeks_price"]
temp_Y_dict = {}
for stock, d in temp_Y.groupby('stock'):
    temp_Y_dict[stock] = d["next_weeks_close"]
```

```python
merged_Y = pd.DataFrame(Y_dict)
Y_final=merged_Y.idxmax(axis=1)
```

```python
test_indexes = Y_final.sample(frac = 0.2,random_state=42).index

X_train={}
X_test={}
y_train={}
y_test ={}
finalY_train= Y_final.drop(test_indexes)
finalY_test= Y_final[test_indexes]


for stock in stock_list:
    X_train[stock] = data_by_stock[stock].drop(test_indexes)
    X_test[stock] = data_by_stock[stock].loc[test_indexes,:]
    y_train[stock] =Y_dict[stock].drop(test_indexes)
    y_test[stock] = Y_dict[stock][test_indexes]
```

The final goal of the problem is to select, every week, the stock that will have the greatest price increase. As such the final labels will be the company with the highest price increase. We will explore the usage of the weeks closing price and the weekly price change as labels for a multi-stage approach.

The prediction will be based only on the pevious week entry as we only have access to 25 weeks of data. It is important to note that to avoid target leakage features not available at the start of the week will not be used and that 2 features have missing values that are filled in.

As for the train test split we will use a 80-20 split, i.e. test set consists of 5 randomly selected weeks.

```python
#add here your modelling code
from sklearn.svm import SVR
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error , accuracy_score , mean_absolute_error
from sklearn.model_selection import cross_validate, RandomizedSearchCV
from scipy.stats import uniform
```

```python
numeric_features = data.select_dtypes(include=np.number).columns.tolist()
categorical_features = data.select_dtypes(exclude=np.number).columns.tolist()

numeric_transformer = Pipeline([
    ('scaler', StandardScaler())])# normalize data

#categorical_transformer = Pipeline(())

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)], remainder='passthrough')

svr_pipeline = {}
for stock in stock_list:
    svr_pipeline[stock] = Pipeline([
    ('preprocessor', preprocessor), 
    ('clf', SVR(C=1.0, epsilon=0.2))])
```

```python
distributions = dict(clf__C=uniform(loc=0, scale=4), clf__gamma=['scale','auto'], clf__epsilon=uniform(loc=0, scale=1))

search = {}
for stock in stock_list:
    clf = RandomizedSearchCV(svr_pipeline[stock], distributions, random_state=0,n_iter=200)
    search[stock] = clf.fit(X_train[stock], y_train[stock])
    svr_pipeline[stock].set_params(**search[stock].best_params_)
```

```python
r2_scores = {}
avg_r2_scores = {}
for stock in stock_list:
    r2_scores[stock] = cross_validate(svr_pipeline[stock], X_train[stock], y_train[stock])['test_score']
    avg_r2_scores[stock] = np.average(r2_scores[stock])
```

```python
for stock in stock_list:
    svr_pipeline[stock].fit(X_train[stock],y_train[stock])
```

```python
def predict_company(pipe_dict, X, stock_list):
    predicted_stock_changes = {}
    for stock in stock_list:
        predicted_stock_changes[stock]=pipe_dict[stock].predict(X[stock])
    predicted_stock_changes_df = pd.DataFrame(predicted_stock_changes)
    predicted_stock_changes_df['date'] = list(X.values())[0].index
    predicted_stock_changes_df = predicted_stock_changes_df.set_index('date')
    
    return predicted_stock_changes_df.idxmax(axis=1)
```

### Train Predictions

```python
pred_train = predict_company(svr_pipeline, X_train, stock_list)
```

### Test Predictions

```python
pred_test = predict_company(svr_pipeline, X_test, stock_list)
```

### New Metric 

```python
def decision_score(best_stock_gt, best_stock_pred , stock_delta):
    pred_result = 0
    gt_result = 0
    for i in range(len(best_stock_pred)):
        pred_result += stock_delta[best_stock_pred[i]].iloc[i]
        gt_result += stock_delta[best_stock_gt[i]].iloc[i]
    return pred_result,gt_result
```

To evaluate the solution three metrics were used: r2(to compare models), accuracy(to evaluate the number of optimal decisions) and the summation of the changes in value(to evaluate the profit made).

We selected a SVR due the algorithm's flexibility. Other models that are expected to perform well in this scenario are lstm (only if more data is made available) and gradient boosting (dropped due to issues with installation).

The approach selected works in two stages, the first predicts this week's stock value variation using a SVR model for each company, the second takes these predictions and selects the stock that has the best predicted value increase. Dividing the approach into two steps has two main advantages: enables the creation of a model for each company, maximises profits based on the predictions.

The model selected achieved a 3% profit in the testing dataset, which is clearly better than the random approach, which has an expected value of -0.45.

New features allow the incorporation of new information improving existing models, while more data points would allow more training and the usage of more complex models.

The solution proposed has several advantages and disadvantages. The main advantages are an individual model per company, minimization of the risk by always selecting the highest price change and simplicity to change companies (when one company is no longer considered for the dow jones all that required is to create a model for the new company). The main disadvantage is that an overestimation of the change cause large damages if the estimation is to far off.

To improve current models the search parameters can be widdened to obtain the best hyper-parameters for each company. Another improvement would be to artificially augment the data creating more data, e.g. training a general model for all companies and afterwards specialising in just one. I would also suggest experimenting with gradient boosting rather than SVR to evaluate if the model achieves better results. Finally, adding a confidence score to the price change predictions could lead to a better second stage stock selection, by taking the confidence score into consideration a risk threshold could be created and adjusted to fit the clients preferences.
